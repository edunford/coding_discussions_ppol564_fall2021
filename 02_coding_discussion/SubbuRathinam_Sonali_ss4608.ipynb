```python
#Importing the dataset for the assignment
import csv
with open('us-states.csv') as file:
    state_covid_data = []
    for row in csv.reader(file):
        state_covid_data.append(row)

len(state_covid_data) #It will display the number of objects in the list
```

Q1) Count up the number of unique dates in the data.

```python
#Creating an empty list to store all the dates in the dataset
date_list = [] 

#For loop to iterate through the dataset. The loop will run from the second row of the dataset, as it is specified in the 
#for loop. This is done as the first row in the dataset is the column heading. 
for i in state_covid_data[1:]:
    if(i[0]):   #Checking if it is the date column in each list
        date_list.append(i[0]) #Adding the date to empty list created earlier
        
print(date_list) #Printing the dates to ensure that the desired output is produced
```

```python
#The set function is used on the date_list list. The set function only takes in distinct values of the parameter passed to it.
date_set = set(date_list) 
#Now that we have a list with distinct dates, we can check the count of unique values using len() function
len(date_set) 
```

Q2) Find the first date in which the District of Columbia recorded a case.

```python
#For loop to find the first date that DC recorded a covid positive case. Again, the for loop won't iterate over the column 
#headings as the range to iterate over has been specified appropriately. 
for i in state_covid_data[1:]:
    if(i[1] == "District of Columbia"): #checking if the state is DC
       print("First date DC recorded a covid case: ", i[0])
       break #When the loop reaches this command, it will terminate. This is implemented here as we only require the first date
    

        
```

Q3) Write a function that takes in a state name as input (e.g. "Wisconsin") and outputs the date of its first case.

```python
"""
state_firstdate function is implemented to display the first recorded covid case in a state. The function has a for loop,
which would iterate over the dataset to find the first recorded case for the state name provided by the user. The result 
is displayed within the loop.

--------------------------------------------------------------------------------------------------------------------------
Arguments: 
state_name -> This will take the state name that has been provided by the user. The default value has been set to None.

---------------------------------------------------------------------------------------------------------------------------
Return Value:
None

---------------------------------------------------------------------------------------------------------------------------
"""

def state_firstdate(state_name=None):

    for i in state_covid_data[1:]:    #Range specified to ensure that column heading is not included
        if(i[1] == state_name):       #Checking if it is required state name 
           print("Date of first recorded covid case in", state_name, " : ", i[0])
           break                      #Only the first case is required, after this statement, the loop will terminate and command moves to the line after for loop 

user_state = input("Enter state name")#User's input

state_firstdate(user_state)           #Function call 
```

```python
import csv
import pandas as pd

file_name = 'coding_discussions_ppol564_fall2021/02_coding_discussion/us-states.csv'
with open(file_name) as file:
    state_covid_data = []
    for row in csv.reader(file):
        state_covid_data.append(row)

len(state_covid_data)
state_covid_data[:5]
state_covid_data
```

### Question 1
### (1) Count up the number of _unique_ dates in the data. 


```python
dates = []
for i in state_covid_data[1:]:
    dates.append(i[0])
    
print("The number of unique dates are:", len(set(dates)))
```

### Question 2
### Find the first date in which the District of Columbia recorded a case. 

```python
dc = []
for i in state_covid_data[1:]:
    if i[1] == "District of Columbia":
        dc.append(i)
        
print("The first date in which DC recorded a case is:", dc[0][0])
```

### Question 3
### Write a function that takes in a _state name_ as input (e.g. "Wisconsin") and outputs the date of its first case.

```python
def take_name(name):
    """ This function takes a name and outputs the date of its first case"""
    data = []
    try:
        for i in state_covid_data[1:]:
            if i[1] == name:
                data.append(i)
        print("The first date in which", name, "recorded a case is:", data[0][0])
    except:
        print("Your entry is not a state or is not spelled correctly. Try again.")
    
take_name("Wisconsin")
```

### Bonus

```python
def when_thousand(name):
    """ This function returns when the total number of cases in a state exceeds a 1000 cases"""
    try:
        data = []
        count = 0
        for i in state_covid_data[1:]:
            if i[1] == name:
                data.append(i)
        for d in data:
            count = count + int(d[3])

            if(count > 1000):
                print("Thousand case in",name,"on :", d[0])
                return

        print("There have not been 1000 cases in this state.")
    except:
        print("Your entry is not a state or is not spelled correctly. Try again.")
        
when_thousand("Wisconsin")
```

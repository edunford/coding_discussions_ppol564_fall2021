```python
#Set up
import pandas as pd

#Import Dataset 
dta = pd.read_csv("~/Desktop/chicago_summer_2018_crime_data.csv")
```

```python
#Create a new dataset which is grouped by the type of crime and day of the month, and add a new a new column which is filled by the number of crimes on that day.  
new_dta=dta.groupby(['primary_type','day']).size().reset_index(name = 'n')
new_dta
```

```python
#create a pivot table.
#We set the type of crime as the index, day of the month as the column, and the number of crimes of each day filled as listed value
crime_proportion = new_dta.pivot_table('amount', index='primary_type', columns='day')
#All missing values are filled with 0.
crime_proportion = crime_proportion.fillna(0)
#Create a new column for the total number of crimes.
crime_proportion = crime_proportion.assign(total_crime = crime_proportion.sum(axis = 1))
crime_proportion
```

```python
#Sort through all cells
for x in range(len(crime_proportion)):
    #Use .iloc tocall to locations in the dtaframe.
    #Divide crime cases of each day by total crime cases.
    crime_proportion.iloc[x]=(crime_proportion.iloc[x]/crime_proportion.iloc[x,-1]).round(2)
    x += 1
    
crime_proportion
    
```

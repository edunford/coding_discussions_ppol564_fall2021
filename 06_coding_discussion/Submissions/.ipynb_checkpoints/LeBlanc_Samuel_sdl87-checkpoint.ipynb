```python
# Import Modules

# Data Management/Investigation
import pandas as pd
import numpy as np
import missingno as miss
from plotnine import *
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Pre-processing
from sklearn import preprocessing as pp 
from sklearn.compose import ColumnTransformer 

# Splits and CV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold # Cross validation 
from sklearn.model_selection import cross_validate # Cross validation 
from sklearn.model_selection import GridSearchCV # Cross validation + param. tuning.

# Machine learning methods 
from sklearn.linear_model import LinearRegression as LM
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.tree import DecisionTreeRegressor as DTree
from sklearn import tree # For plotting the decision tree rules
from sklearn.ensemble import BaggingRegressor as Bag
from sklearn.ensemble import RandomForestRegressor as RF

# Evaluating performance
import sklearn.metrics as m

# Pipeline
from sklearn.pipeline import Pipeline
```

## Load and Inspect Data

```python
# Load training data for DC Housing Prices
train_dat = pd.read_csv("../train_data.csv")
train_dat.shape
```

```python
# Load test data for DC Housing Prices
test_dat = pd.read_csv("../test_data.csv")
test_dat.shape
```

```python
# Look for missing training data
miss.matrix(train_dat)
```

## Choose Predictors and Visualize

```python
# Define list of predictors to predict the log sale price of a DC house
predictors = ["BATHRM", "ROOMS", "BEDRM", "FIREPLACES","KITCHENS","GRADE"]
```

```python
# Drop missing observations for our predictors
train_dat = train_dat.dropna(subset = predictors)
test_dat = test_dat.dropna(subset = predictors)

# Drop observations where the outcome is missing
train_dat = train_dat[~train_dat.LN_PRICE.isna()]
test_dat = test_dat[~test_dat.LN_PRICE.isna()]

# Only consider houses that actually sold (i.e. sale price > 0)
train_dat = train_dat[train_dat.LN_PRICE > 0]
test_dat = test_dat[test_dat.LN_PRICE > 0]
```

```python
# Split into training and test data based on predictors
y_train = train_dat.LN_PRICE
x_train = train_dat[predictors]

y_test = test_dat.LN_PRICE
x_test = test_dat[predictors]
```

```python
# Explore the training data to see how the predictors are distributed
(
    ggplot(x_train.melt(), aes(x = "value")) +
    geom_histogram(binwidth = 0.5) +
    facet_wrap("variable", scales = "free") +
    theme(figure_size = (10, 3))
)
```

```python
# Explore the outcome (LN_PRICE) to check for issues
(
    ggplot(pd.DataFrame(dict(LN_PRICE = y_train)), aes(x = "LN_PRICE")) +
    geom_histogram(binwidth = 0.1) +
    theme(figure_size = (7, 3)) 
)
```

```python
# Explore how the predictors relate to the outcome LN_PRICE

# Create copy of the training x-values
X_ = x_train.copy()

# Combine with the outcomes from the training y-values
X_["LN_PRICE"] = y_train

# Plot the predictors against the outcome
(
    ggplot(X_.melt(id_vars = ["LN_PRICE"]), aes(x = "value", y = "LN_PRICE")) +
    geom_point(alpha = .02) +
    facet_wrap("variable", scales = "free") +
    geom_smooth(method = "lm", se = False, color = "red") +
    theme(figure_size = (10, 3)) 
)
```

### We can see that there is a positive relationships between each of our predictorss and the outcome variable.

## Create and Run Model

```python
# Create folds to ensure comparable samples
fold_generator = KFold(n_splits = 5, shuffle = True, random_state = 12321)
```

```python
# Preprocess the predictors using a MinMaxScaler
preprocess = ColumnTransformer(transformers = [("num", pp.MinMaxScaler(), predictors)])
```

```python
# Create our model Pipeline,note that we leave the model as None for now to act as a placeholder
pipe = Pipeline(steps=[("pre_process", preprocess),("model", None)])
```

```python
# Specify the five models to run (and their respective tuning parameters)
search_space = [
    # Linear Model
    {"model" : [LM()]},
    
    # KNN with K tuning param
    {"model" : [KNN()],
     "model__n_neighbors" : [10, 20, 40, 80, 160]},
    
    # Decision Tree with the Max Depth param
    {"model" : [DTree()],
     "model__max_depth" : [1, 2, 3, 4, 5]},
    
    # The Bagging decision tree model 
    {"model" : [Bag()]},
    
    # Random forest with the N Estimators and Max Depth tuning params
    {"model" : [RF()],
     "model__max_depth" : [1, 2, 3, 4, 5],
     "model__n_estimators" : [500, 1000, 1500, 2000]}
]
```

```python
# Combine it all together with GridSearch
search = GridSearchCV(pipe, 
                      search_space, 
                      cv = fold_generator,
                      scoring = "neg_mean_squared_error",
                      n_jobs = 5)
```

```python
# Fit the model to the training data
search.fit(x_train, y_train)
```

```python
# Identify the best model out of the search
search.best_params_
```

```python
# Use the predict method with best model from the search
y_pred = search.predict(x_test)
```

## Evaluate Model

```python
# Get the mean squared error
m.mean_squared_error(y_test, y_pred)
```

```python
# Get the r-squared value
m.r2_score(y_test, y_pred)
```

```python
# Plot predicted values vs. true values of outcome variable to observe how well the model performs
(
    ggplot(pd.DataFrame(dict(pred = y_pred, true = y_test)), aes(x = "pred", y = "true")) +
    ggtitle("Best Model Preformance") +
    xlab("Prediction Value") +
    ylab("True Value") +
    xlim(12, 16) +
    ylim(0, 20) +
    geom_point(alpha = .1) +
    geom_abline(color = "red", size = 1) +
    theme(figure_size = (10, 10))
)
```

### The model that best fit our data with the lowest mean squared error was the K-nearest neightbors model with the parameter of 160 neighbprs. The model we chose was the one with the highest number of neighbors in our test set, so we note that a model of significantly more than 160 neighbors may still be a better model. The final R^2 was 0.44, with a mean squared error of 0.41.

```python
#Import packages
import numpy as np
import pandas as pd
import re
```

```python
#Read the text files
with open('aljazeera-khashoggi.txt') as f:
    aljazeera=f.read()
with open('bbc-khashoggi.txt') as f:    
    bbc=f.read()
with open('breitbart-khashoggi.txt') as f:  
    breitbart=f.read() 
with open('cnn-khashoggi.txt') as f:   
    cnn=f.read()
with open('fox-khashoggi.txt') as f:   
    fox=f.read()
```

```python
#Check one file
aljazeera
```

```python
#Read the stopwords and transfer it into list form
stopword=pd.read_csv('stop_words.csv')
stopword = list(stopword["word"])
```

## Analysis on correlation by removing stopwords

```python
#Turn the text into numerical data.
#Break up the sentences into individual words, scrub capitalization punctuation and numbers
def tokenize(text=None):
    '''
    Turn the text into numerical data.
    
    Parameters
    ------
    text: a string of txt file's name.
    
    Returns
    ------
    text_list: a list that contains each word of the txt file.
    '''
    text = text.lower()
    text = re.sub(r'[^\w\s]','',text)#Remove the punctuations
    text = re.sub(r'[0-9]+', '', text) #Remove the numbers
    text_list = text.split()#Split the word
    text_list = [word for word in text_list if word not in stopword]#Remove word in stopwords
    return text_list

#Test file 'aljazeera'
tokenize(aljazeera)
```

```python
#Counts the number of words in the text
def wordsnumber(txt=None):
    '''
    Counts the number of words in the text
    
    Parameters
    ------
    txt: a string of txt file's name.
    
    Returns
    ------
    wordsnumber: a list that contains the number of each word in the txt file.
    '''
    #Create a dictionary
    d = dict()
    #Caculate the number of each word
    for word in tokenize(txt):
        if word in d:
            d[word][0] += 1
        else:
            d[word] = [1]
    return pd.DataFrame(d)
#Test file 'aljazeera'
wordsnumber(aljazeera)
```

```python
def gen_DTM(texts=None):
    '''
    Generate a document term matrix
    
    Parameters
    ------
    texts: a string of txt file's name.
    
    Returns
    ------
    DTM: a dataframe that contains the number of each word for all txt file.
    '''
    DTM = pd.DataFrame()
    for text in texts:
        entry = wordsnumber(text)
        DTM = DTM.append(pd.DataFrame(entry),ignore_index=True,sort=True) # Bind rows
    
    DTM.fillna(0, inplace=True) # Fill in  missing values with 0s 
    return DTM

# Test for all text files      
gen_DTM([aljazeera, bbc, breitbart, cnn, fox])
```

```python
#Calculate the angle between two vectors
def cosine(a,b):
    '''
    Generate a function to calculate the angle between two vectors
    
    Parameters
    ------
    a,b: float 
         two vectors
    
    Returns
    ------
    cos: float
         the cosine value of the angle between the two vectors
    '''
    cos = np.dot(a,b)/(np.sqrt(np.dot(a,a)) * np.sqrt(np.dot(b,b))  )
    return cos

```

```python
#Index the pandas dataframe to draw out a vector
D = gen_DTM([aljazeera, bbc, breitbart, cnn, fox])
a = D.iloc[0].values
b = D.iloc[1].values
c = D.iloc[2].values
d = D.iloc[3].values
e = D.iloc[4].values
```

```python
#Create a new dictionary to assign the corresponding name for each vector 
df_nostop = {
'Aljazeera':a,
'BBC':b,
'Breitbart':c,
'CNN':d,
'Fox':e
} 
#Transfer the dictionary into dataframe form
df_nostop = pd.DataFrame(df_nostop)
#Make cosine correlation dataframe for the vectors
df_nostop_corr = df_nostop.corr(method=cosine).round(4)
#Dataframe for the correlation betetween vectors(not includes stopwords)
df_nostop_corr
```

## Analysis on correlation including stopwords

```python
#Turn the text into numerical data.
#Break up the sentences into individual words, scrub capitalization punctuation and numbers
def tokenize2(text=None):
    '''
    Turn the text into numerical data.
    
    Parameters
    ------
    text: a string of txt file's name.
    
    Returns
    ------
    text_list: a list that contains each word of the txt file.
    '''
    text = text.lower()
    text = re.sub(r'[^\w\s]','',text)#Remove the punctuations
    text = re.sub(r'[0-9]+', '', text) #Remove the numbers
    text_list2 = text.split()#Split the word
    return text_list2

#Test file 'aljazeera'
tokenize2(aljazeera)
```

```python
#Counts the number of words in the text
def wordsnumber2(txt=None):
    '''
    Counts the number of words in the text
    
    Parameters
    ------
    txt: a string of txt file's name.
    
    Returns
    ------
    wordsnumber2: a list that contains the number of each word in the txt file.
    '''
    #Create a dictionary
    d = dict()
    #Caculate the number of words
    for word in tokenize2(txt):
        if word in d:
            d[word][0] += 1
        else:
            d[word] = [1]
    return pd.DataFrame(d)

wordsnumber2(aljazeera)
```

```python
def gen_DTM2(texts=None):
    '''
    Generate a document term matrix
    
    Parameters
    ------
    texts: a string of txt file's name.
    
    Returns
    ------
    DTM2: a dataframe that contains the number of each word for all txt file.
    '''
    DTM2 = pd.DataFrame()
    for text in texts:
        entry = wordsnumber2(text)
        DTM2 = DTM2.append(pd.DataFrame(entry),ignore_index=True,sort=True) # Row bind
    
    DTM2.fillna(0, inplace=True) # Fill in any missing values with 0s (i.e. when a word is in one text but not another)
    return DTM2

# Test for all text files      
gen_DTM2([aljazeera, bbc, breitbart, cnn, fox])

```

```python
#Index the pandas dataframe to draw out a vector
D2 = gen_DTM2([aljazeera, bbc, breitbart, cnn, fox])
a2 = D2.iloc[0].values
b2 = D2.iloc[1].values
c2 = D2.iloc[2].values
d2 = D2.iloc[3].values
e2 = D2.iloc[4].values
```

```python
#Create a new dictionary to assign the corresponding name for each vector 
df_stop = {
'Aljazeera':a2,
'BBC':b2,
'Breitbart':c2,
'CNN':d2,
'Fox':e2
} 
#Transfer the dictionary into dataframe form
df_stop = pd.DataFrame(df_stop)
#Make cosine correlation dataframe for the vectors
df_stop_corr = df_stop.corr(method=cosine).round(4)
#Dataframe for the correlation betetween vectors(includes stopwords in the text)
df_stop_corr
```

```python
#Dataframe for the correlation betetween vectors(not includes stopwords in the text)
df_nostop_corr
```

# Conclusion
From the two correlation(similarity) tables, compared table with stopwords removed, the correlation between these reports is higher when including stopwords.
Case1(including stopwords in text):The report text of BBC and Breitbart, of BBC and Fox are extremly high(cosine value at 0.899 and 0.89 respectively). CNN's content and Breitbart's content has much difference, the cosine value is just 0.682.
Case2(removing stopwords in text): Fox and Aljazeera has highest similarities with cosine value at 0.698. However, the report text of BBC and Fox is not much correlated at this time. Also, CNN and Breitbart has the least similarity with cosine value at 0.378.
We can see that when we include stopwords in analysis of correlation, some information we get might be misleading.
We may get a high correlation result because of the high frequency use of stopwords, which are common to the English Language but make little sense of the main purpose of the text.

```python
data = pd.read_csv('Data\stop_words.csv')
 
# converting column data to list
stopwords = data['word'].tolist()
print(stopwords)
```

```python
# Follwoing token function is used to clean out stopwords words
def tokenize(text=None):
    '''
    Tonkenizes the text and breaks up into single words and removes stopwords
    '''
    text = text.lower()
    text = text.replace('.','')
    text_list = text.split()
    text_list2 = [word for word in text_list if word not in stopwords]
    return text_list2
```

```python
def convert_text_to_dtm(txt):
    '''
    Converts text into a document term matrix.
    '''
    d = dict()
    for word in tokenize(txt):
        if word in d:
            d[word][0] += 1
        else:
            d[word] = [1]
    return pd.DataFrame(d)

convert_text_to_dtm(news1)
```

```python
def gen_DTM(texts=None):
    '''
    Generate a document term matrix
    '''
    DTM = pd.DataFrame()
    for text in texts:
        entry = convert_text_to_dtm(text)
        DTM = DTM.append(pd.DataFrame(entry),ignore_index=True,sort=True) # Row bind
    
    DTM.fillna(0, inplace=True) # Fill in any missing values with 0s (i.e. when a word is in one text but not another)
    return DTM 
```

```python
def cosine(a,b):
    cos = np.dot(a,b)/(np.sqrt(np.dot(a,a)) * np.sqrt(np.dot(b,b))  )
    return cos
```

```python
D = gen_DTM([news1, news2, news3, news4, news5])
aljazeera = D.iloc[0].values
bbc = D.iloc[1].values
breitbart = D.iloc[2].values
cnn = D.iloc[3].values
fox = D.iloc[4].values
cosine(aljazeera, bbc) 
```

```python
# Check for Breitbart and Fox
cosine(breitbart, fox) 
```

```python
matrix = {'aljazeera': aljazeera,
          'bbc': bbc,
          'breitbart': breitbart,
          'cnn': cnn,
          'fox': fox}


matrixdf = pd.DataFrame(matrix)

newscorr = matrixdf.corr(method=cosine)

newscorr
```

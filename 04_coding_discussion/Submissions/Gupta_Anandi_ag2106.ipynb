```python
#basic cleaning and splitting of text

def tokenize(text=None):
    text = text.lower()
    text = text.replace('.','')
    text_list = text.split()
    return text_list
```

```python
def convert_text_to_dtm(txt):
    '''
    Converts text into a document term matrix.
    '''
    d = dict()
    for word in tokenize(txt):
        if word in d:
            d[word][0] += 1
        else:
            d[word] = [1]
    return pd.DataFrame(d)

convert_text_to_dtm(cnn)
```

```python
# Now build a function that does this for a list of texts
def gen_DTM(texts=None):
    '''
    Generate a document term matrix
    '''
    DTM = pd.DataFrame()
    for text in texts:
        entry = convert_text_to_dtm(text)
        DTM = DTM.append(pd.DataFrame(entry),ignore_index=True,sort=True) # Row bind
    
    DTM.fillna(0, inplace=True) # Fill in any missing values with 0s (i.e. when a word is in one text but not another)
    return DTM

#apply to the five news articles     
gen_DTM([aljazeera, bbc, breitbart, fox, cnn]) 
```

```python
#calculate cosines using numpy

def cosine(a,b):
    cos = np.dot(a,b)/(np.sqrt(np.dot(a,a)) * np.sqrt(np.dot(b,b))  )
    return cos
```

```python
#generate document term matrix for all five articles and save to dictionary

def calc_corr():
    

    D = gen_DTM([aljazeera, bbc, breitbart, fox,cnn]) 
    data = {'aljazeera': D.iloc[0].values,
        'bbc': D.iloc[1].values,
        'breitbart': D.iloc[2].values,
        'fox': D.iloc[3].values,
        'cnn': D.iloc[4].values
        }

    df = pd.DataFrame(data,columns=['aljazeera','bbc','breitbart','fox','cnn'])

#calculate cosines for all combinations of news sources

    corrMatrix = df.corr(method = cosine)
    return (corrMatrix)
    
```

```python
#redefine cleaning function to only keep alphabets

def tokenize(text=None):
    text = text.lower()
    text = text.replace('.','')
    text_list = text.split()
    text_list2 = [''.join(e for e in string if e.isalpha()) for string in text_list]
    text_list3 = [word for word in text_list2 if word not in stop_words]
    return text_list3
```

```python
#recalculate cosines after removing numbers and special characters

calc_corr()
```

```python
#find max and min cosines
c = calc_corr().abs()

s = c.unstack()
so = s.sort_values(kind="quicksort")
so
```

In general, there seems to be a fair amount of similarity between how the Khashoggi scandal was reported across the five different sources. Aljazeera and Fox seem to be the most similar (cosine of 0.71 after removing stop words, numbers, and special characters) and Breitbart and CNN seem to be the least similar (cosine of 0.38 after removing stop words, numbers, and special characters). 

Cosines are highest with minimal cleaning (before removing stop words, numbers, or special characters). This is because all of the news sources likely use words that are common to most sentences, and this is inflating how similar the sources appear. Removing stop words lower the cosines substantially. However, once we remove numbers and other special characters, we see cosines increase slightly. This also makes intuitive sense, as these characters are generally meaningless, but differences in how numbers and characters are used by different authors will make the sources appear less similar to each other than they truly are.

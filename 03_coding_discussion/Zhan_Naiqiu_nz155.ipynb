```python
# Count the case amount of every day, group by type. 
chicago_crime1 = (chicago_crime
                  .filter(['primary_type', 'month', 'day'])
                  .groupby(['primary_type','month', 'day'])
                  .size()
                  .reset_index(name = 'Case of day'))

chicago_crime1
```

```python
# Create a pivot table for case number of every type in each day.
chicago_final = chicago_crime1.pivot_table(index = ['primary_type', 'month'], columns = ['day'], fill_value = 0)
chicago_final.head()
```

```python
# Count the total of every month
chicago_final.loc[:,'Total'] = chicago_final.sum(axis=1)
chicago_final.head()
```

```python
# Calculate the rate use this loop. 
for i in range (len(chicago_final)):
    chicago_final.iloc[i] = (chicago_final.iloc[i]/chicago_final.iloc[i, -1]).round(2)
    
chicago_final   
```

```python
# Change the pivot table to records format. 
c_rate = pd.DataFrame(chicago_final.to_records())
c_rate
```

```python
#import pandas and the data
import pandas as pd
chi_crime = pd.read_csv("chicago_summer_2018_crime_data.csv")

#take a look at the data to get a sense of it.
chi_crime.head()
```

```python
#Create a subset of the original dataframe leaving only the type and the day. Calculate the sum by crime type and day.
chi_crime_subset = chi_crime.groupby(["primary_type","day", "month"]).size().reset_index(name = "sum")

#take a look at it.
chi_crime_subset.head()
```

```python
#transform that subset of data into wide format.
chi_crime_long = chi_crime_subset.pivot_table(values='sum',columns='day', index=['primary_type', "month"], fill_value = 0)

#take a look at it.
chi_crime_long.head()
```

```python
#calculate the sum of numbers of each crime throughout the days and assign them to a new column.
chi_crime_long = chi_crime_long.assign(sum_crime = chi_crime_long.sum(axis=1))

#Again, take a look at it.
chi_crime_long.head()
```

```python
#iterate through each cell to calculate the percentage.
for i in range(len(chi_crime_long)):
    for j in range(len(chi_crime_long.columns) - 1):
        chi_crime_long.iloc[i, j] = round(chi_crime_long.iloc[i, j]/chi_crime_long.iloc[i, -1], 2)

#show the whole data frame.
chi_crime_long   
```

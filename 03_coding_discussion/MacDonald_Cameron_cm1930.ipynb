```python
# load packages
import pandas as panda
```

#### Method # 1

```python
# import data
crime_data = panda.read_csv("chicago_summer_2018_crime_data.csv")
# show the table
crime_data.head()
```

```python
# remove unneeded columns
crime_data = crime_data[['day', 'primary_type']]
# create a frequency table with the crime type and the day
crime_data = panda.crosstab(crime_data['primary_type'],crime_data['day'])
```

```python
# divide values by the row sum
crime_data = crime_data.div(crime_data.sum(axis = 1), axis = 0) 
# round to 2 decimal places
crime_data = crime_data.round(decimals = 2)
```

#### Method # 2

```python
# import data for the second time
crime_data2 = panda.read_csv("chicago_summer_2018_crime_data.csv")
```

```python
# remove unneeded columns
crime_data2 = crime_data2[['day', 'primary_type']]
# check out data 
crime_data2.head()
```

```python
# group the data by the type and day, get count of each type by day, unstack (it is a list object), remove NaN so that we can make everything an integer, make ints
crime_data2 = crime_data2.groupby(['primary_type', 'day']).size().unstack().fillna(0).astype(int)
# divide each value by row sum
crime_data2 = crime_data2.div(crime_data2.sum(axis = 1), axis = 0)
# round
crime_data2 = crime_data2.round(decimals = 2)
```

```python
#Group by event type and date for subtotals for each day
x = data_chicago.groupby(['primary_type', 'day']).size().reset_index(name = "counts")

#Group by event type for totals for each event type
y = data_chicago.groupby(['primary_type']).size().reset_index(name = "total")

#Merge totals on to subtotals
final = x.merge(y, on='primary_type', how='left')

#Calculate the proportions
final['proportion'] = round(final['counts']/final['total'], 2)

#Convert dataset from long to wide
final_wide = final.pivot_table(values='proportion', columns='day', index='primary_type')
```

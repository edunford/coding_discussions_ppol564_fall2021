```python
#importing pandas module
import pandas as pd 

#importing Chicago CSV and assinging it the name "dat"
dat=pd.read_csv("/Users/ellisobrien/Desktop/Georgetown Semester 1/Data Science/coding_discussions_ppol564_fall2021/03_coding_discussion/chicago_summer_2018_crime_data.csv")

```

```python

#creating data frame and grouping by crime type and day
days = dat.groupby(['primary_type','day']).size().reset_index(name = 'crimecount_byday')

days[:5]



```

```python
#creating data frame group just by type 
crime= dat.groupby(['primary_type']).size().reset_index(name = 'crimecount_all')

crime[:5]

```

```python
#joining two tables with right join
crime_joined = crime.merge(days, on='primary_type', how='right')

#creating new column with crime rate by day 
crime_joined['crime_rate'] = round(crime_joined['crimecount_byday']/crime_joined['crimecount_all'], 2)

crime_joined [:5]
```

```python
#pivoting crime rate values so that days are columns 
crime_pivoted=crime_joined.pivot_table(values = 'crime_rate', index = 'primary_type', columns = 'day')


#adding 0's as NA's 
crime_pivoted.fillna(0.00)
```

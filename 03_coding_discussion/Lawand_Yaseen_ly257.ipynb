Requirements: 
1. Unit of observation is the crime type (primary type)
2. Column variables correspond with the day of the month 
3. Each cell is populated with proportion of times that crime type was committed over all days of the month


```python
import pandas as pd 
```

```python
#importing raw data 

chicago_crime = pd.read_csv("/Users/lawandyaseen/Desktop/coding_discussions_ppol564_fall2021/03_coding_discussion/chicago_summer_2018_crime_data.csv")

#chicago_crime
```

```python
#grouping data by the day of instance and by the type of crime listed 
#.size returns the number of instances for each row and adds it as new column 
#naming column "amount"

crime_by_type_date = chicago_crime.groupby(['day', 'primary_type']).size().reset_index(name = 'amount')

#crime_by_type_date
```

```python
#using pivot table to go from long to wide 
#row - type of crime 
#column - day of the month 
#any values that are missing are equal to 0 
proportion_crime = crime_by_type_date.pivot_table(columns = 'day', index = 'primary_type', fill_value = 0)

#creating new column called total which sums the instances of specific crime over the course of the month 
proportion_crime = proportion_crime.assign(total = proportion_crime.sum(axis = 1))

#proportion_crime
```

```python
#iterator dummy value 
x = 0

#using iloc to iterate through each row to divide the instances on specific day and divide it by the total instances column
#also rounding to two decimal places
for x in range(len(proportion_crime)):
    proportion_crime.iloc[x] = (proportion_crime.iloc[x]/ proportion_crime.iloc[x,31]).round(2)
    x += 1

proportion_crime
```

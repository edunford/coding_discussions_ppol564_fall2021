```python
import pandas as pd
```

```python
df = pd.read_csv("chicago_summer_2018_crime_data.csv")
df.head()
```

```python
# Select subset of data frame with crime type and date columns only
df_type_day = df[["day", "primary_type"]]
df_type_day.head()
```

```python
# Create new column for the count of crimes in a single day of the month
# The count will start at 1
df_type_day = df_type_day.assign(crime_count = 1)
# Print first five rows of data frame to check execution of new column
df_type_day.head()
```

```python
# Pivot data frame from long to wide
# Use "count" function to sum 1s in "crime_count" column
# Replace any missing values with a 0
df_pivot = df_type_day.pivot_table(index = "primary_type", columns = "day", aggfunc = "count", fill_value = 0)
# Print first five rows of data frame to check execution of pivot and count function
df_pivot.head()
```

```python
# Calculate proportions for each cell value
df_proportions = (df_pivot
                      # Divides each value by the sum of the row it's in
                      .div(df_pivot.sum(axis = 1), axis = 0)
                      # Round each number to 2 decimal places
                      .round(2))
# Print entire data frame
df_proportions
```

#### 1 : Import Libraries, set WD, and load file

```python
import pandas as pd, os
path = r'C:\Users\unily\Documents\Georgetown\PPOL 564 - Intro to Data Science\Coding discussions'
os.chdir(path) # Set WD
df_1 = pd.read_csv("chicago_summer_2018_crime_data.csv")[["primary_type", "day"]] #Load file, variables of interest only
df_1.sample(3) # Print a random draw of 3 observations
```

#### 2: Create new DF as intructed

```python
# 2.1: Use grouby to create new DF with proportion of times crime types were committed over all days of the month.
df_2 =  (df_1.groupby(df_1.day).primary_type.value_counts() / df_1.groupby(df_1.primary_type).size()).\
                                                                round(2).to_frame(name='proportion')
# 2.2: Pivot 'df_2' to wide, leaving primary_type as unit of observation, and replace missing values with 0.
new_df = df_2.pivot_table(values = 'proportion', index = 'primary_type', columns= 'day',fill_value=0)
new_df #Printing new data frame
```

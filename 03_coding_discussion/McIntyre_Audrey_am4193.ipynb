```python
import pandas as pd

crimedat = pd.read_csv("chicago_summer_2018_crime_data.csv")
crimedat.head()
```

```python
#finding the count of each type of crime on each day
byday = (crimedat
         .groupby(['primary_type','day'])
         .size()
         .reset_index(name="n")
        )
byday
```

```python
#pivot the data with day counts to have crime type as rows and days as columns
newcrime = byday.pivot_table(byday, index= 'primary_type', columns = 'day', fill_value = 0)
#add a column summing across columns for the total of that type 
newcrime['sum']= newcrime.sum(axis=1)
newcrime
```

```python
#Divide each day's count by the total for that type
for c in range(len(newcrime)): 
    newcrime.iloc[c] = (newcrime.iloc[c]/newcrime.iloc[c,31]).round(2)

newcrime
```

```python
# Import pandas library
import pandas as pd
```

```python
# Import data and look at first couple rows
initial_data = pd.read_csv("chicago_summer_2018_crime_data.csv")
initial_data.head()
```

```python
# Create a new data frame containing the day, primary type, and number of occurances 
sorted_by_day = initial_data.groupby(['day', 'primary_type']).size().reset_index(name = 'count')
sorted_by_day.head()
```

```python
# Cast the data to a wide format with a pivot table
sorted_pivot = sorted_by_day.pivot_table(index = "primary_type", columns = "day", fill_value = 0)
sorted_pivot.head()
```

```python
# Add a new column containing the sum across each row
sorted_pivot['total'] = sorted_pivot.sum(axis=1)
sorted_pivot.head()
```

```python
# Iterate through the table and divide the number of occurances by the total 
for x in range(len(sorted_pivot)):
    sorted_pivot.iloc[x] = round(sorted_pivot.iloc[x]/sorted_pivot.iloc[x, -1], 2)

# Drop the total column to just include the days
final_data = sorted_pivot.drop(columns = 'total')
final_data
```

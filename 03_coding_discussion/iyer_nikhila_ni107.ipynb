```python
# Import the necessary packages
import pandas as pd
import numpy as np
```

```python
# Load in the Chicago Crime - Summer 2018 data set, and check if it loaded correctly by printing a small chunk
chi_crime = pd.read_csv ('/Users/nikhilaiyer/Documents/GRAD SCHOOL/ppol564/coding_discussions_ppol564_fall2021/03_coding_discussion/chicago_summer_2018_crime_data.csv')
chi_crime.head()
```

```python
## ROWS
# Set up the rows for the new data frame: The rows headers are the various types of crime ('primary_type')
# Also get the total number of each type of crime and save it to a variable to use later when calculating proportions
rows = chi_crime.primary_type.unique()
rowSums = chi_crime.groupby("primary_type").size().sort_values(ascending = False)
```

```python
## COLUMNS
# Set up the columns for the new data frame: The columns are each one day of a month, which is a range from 1-31
cols = pd.Series(range(1,32))
```

```python
## NEW DATA FRAME
# Create the new data frame and set the rows and columns accordingly as indicated above
# The data frame will populate with NaN for the time being, and we will fill it with the appropriate data by cell
newdf = pd.DataFrame(columns = cols, index = rows)
```

```python
## CELL DATA
# Calculate the data for each cell, which will be the percentage / proportion of a certain type of crime (row) that happened on that day (column)
# Going through the newly set up data frame, for every row for every column, find the total number of that crime that happened on that day
# Then, take that data (cell) and divide it by the total number of that crimes that happened in the month
# Populate the appropriate cell with that proportion data point, rounded to the second decimal
for r in rows:
    for c in cols:
        cell = ((chi_crime['primary_type'] == r) & (chi_crime['day'] == c)).sum()
        newdf.loc[r, c] = (cell / rowSums[r]).round(2)
```

```python
## FINAL DATA frame
# Sort the new data frame alphabetically, and print it out!
newdf.sort_index(ascending = True)
```

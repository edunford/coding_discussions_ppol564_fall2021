```python
#import libraries
import pandas as pd
```

```python
#read in the chicago crime data
data = pd.read_csv("chicago_summer_2018_crime_data.csv")
#look at data structure
print(len(data))
data.head()
```

```python
#find total amount of each crime 
totals = pd.DataFrame(data.groupby(['primary_type']).size().reset_index(name = "total_count"))
totals
```

```python
#group by both day and primary type and assign to new dataframe
grouped = pd.DataFrame(data.groupby(['primary_type', 'day']).size().reset_index(name = "daily_count"))
#check new dataframe
grouped
```

```python
#merge two together for easier division
grouped = grouped.merge(totals, how="left", on="primary_type")
grouped
```

```python
#create new column to hold portion of crime committed on any given day
grouped['proportion'] = round(grouped['daily_count']/grouped['total_count'], 2)
grouped
```

```python
#pivot table so each type only has one row
data_by_type = grouped.pivot_table(values='proportion', columns='day', index='primary_type', fill_value=0)
data_by_type
```

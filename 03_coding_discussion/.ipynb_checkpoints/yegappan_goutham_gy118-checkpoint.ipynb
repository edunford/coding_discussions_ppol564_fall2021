```python
import pandas as pd

data = pd.read_csv("coding_discussions_ppol564_fall2021/03_coding_discussion/chicago_summer_2018_crime_data.csv")
data
```

```python
final = pd.DataFrame(index = data['primary_type'].unique(), columns = range(1, 32))\
        .fillna(0.0)
for crime in final.index:
    dt = data.where(data['primary_type'] == crime).dropna()
    number_of_crimes = len(dt)
    
    for date in range(1, 32):
        
        dt2 = dt.where(dt['day'] == date).dropna()
        crimes_on_date = len(dt2)
        prop = crimes_on_date/number_of_crimes
        final.at[crime, date] = round(prop, 2)
display(final)
```

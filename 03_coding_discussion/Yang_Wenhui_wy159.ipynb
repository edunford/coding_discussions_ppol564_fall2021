```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./chicago_summer_2018_crime_data.csv",)
#df = df[['primary_type', 'day']]
# create a new column called "happen" this is used to count the number of occurrence.
df = df.assign(happen=1)
# cast the data with the items we need and sum up values
table = df.pivot_table(values='happen', index=['primary_type'], columns=['day'], aggfunc=np.sum)
# fill all missing values with zeros
table = table.fillna(0)
# convert times of occurrence to relative frequency 
table = table.apply(lambda x: round(x/x.sum(),2),axis = 1)
table
```

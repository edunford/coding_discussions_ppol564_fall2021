```python
#preparation
import pandas as pd
```

```python
#load data & display first 5 rows
crimeData = pd.read_csv("chicago_summer_2018_crime_data.csv")
crimeData.head()
```

```python
#Subset data to only pull the 2 columns we need - Day & Crime Type
crimeDayType = crimeData[["day", "primary_type"]]

#Display first 5 rows
crimeDayType.head()
```

```python
#Create a pivot table of crime type counts per day
tableCrime = pd.pivot_table(crimeDayType, index=['primary_type'], columns=['day'], values=['day'], 
                            aggfunc={'day': 'count',})

#replace NAs with 0 
tableCrime = tableCrime.fillna(0)

#Display first 5 rows
tableCrime.head()
```

```python
#Use lambda function to proportion values across columns
tableCrimeProp = tableCrime.apply(lambda x: x/x.sum(), axis=1).round(2)

#Display
tableCrimeProp
```

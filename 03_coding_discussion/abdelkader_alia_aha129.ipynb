```python
import pandas as pd

# Read in the data
cdata = pd.read_csv('/Users/Alia/Documents/Github/coding_discussions_ppol564_fall2021/03_coding_discussion/chicago_summer_2018_crime_data.csv')

# Group data by crime type and index by day
cgroup = cdata.groupby('primary_type')['day']

# Create data frame with crime type as index and days 1-31 as columns
ret_df = pd.DataFrame(index=cdata["primary_type"].unique(),columns=range(1, 32))

# Loop through tuples of crime type and series of dates in cgroup
for crime_type, crime_data in cgroup:
    # Count total instances of crime
    crime_total = len(crime_data)
    # Create an array with length 31 to count crimes per day
    days_array = [0] * 31
    for day in crime_data:
        # Add 1 for each crime instance
        days_array[day - 1] += 1
    # Normalize daily crime count by total crime count
    days_array_norm = [round(count / crime_total, 2) for count in days_array]
    ret_df.loc[crime_type] = days_array_norm

ret_df
```

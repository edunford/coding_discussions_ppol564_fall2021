```python
# importing pandas
import pandas as pd
# loading csv file
dat = pd.read_csv('chicago_summer_2018_crime_data.csv')
# displaying first 5 observations of data
dat.head()
```

```python
# Create new df with the month, day, and crime type variables from the csv
crime_df = dat[["month","day", "primary_type"]]  
# Add new column to crime_df to count each crime
crime_df["count"] = crime_df.apply(lambda x: [1], axis=1, result_type = "expand" )
# Display first 5 rows of crime_df
crime_df.head()
```

```python
# Pivoting data frame from long to wide, indexing by crime type, summing the counts for each crime and day, and replacing missing values with 0
crime_df_pivot = crime_df.pivot_table(columns = "day", index = "primary_type", aggfunc = "count", fill_value = 0)
# Show first 5 rows of pivoted data frame
crime_df_pivot.head()
```

```python
# Create new data frame with proportions by summing total counts for each crime and month, and dividing by the cell value
crime_proportions_per_day = \
(crime_df_pivot
    .div(crime_df_pivot.sum(axis = 1), axis = 0)
    .round(2))

# Print data frame
crime_proportions_per_day
```

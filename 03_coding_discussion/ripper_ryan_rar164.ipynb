```python
# Import module.
import pandas as pd
```

```python
# Load in the Chicago Summer 2018 Crime Data
crime_data = pd.read_csv("chicago_summer_2018_crime_data.csv")
```

```python
# Check to make sure the data was loaded properly.
crime_data.sample(5)
```

```python
# Create new data frame with the columns of interest, day and primary_type.
crime_data_day_type = crime_data.filter(["day", "primary_type"])
```

```python
# On the above data frame, apply a pivot table to obtain count for each type by day.
# Index is "primary_type", columns is "day", count occurrences, and fill empties with "0".
crime_data_day_type_count = (pd.pivot_table(crime_data_day_type, 
                                index = "primary_type", 
                                columns = "day", 
                                aggfunc = len, 
                                fill_value = 0, 
                                margins = True)
                            )
```

```python
# Utilize the margins results from the pivot table to divide the count across each type to calculate percentages.
crime_data_day_type_perc =  (crime_data_day_type_count
                                .div(crime_data_day_type_count.iloc[:, -1], axis = 0) # Divide count across row.
                                .round(2) # Round to the second decimal place.
                                .iloc[:-1, :-1] # Remove the margins results from the pivot table.
                            )
```

```python
# Print out results from above data wrangling.
crime_data_day_type_perc
```

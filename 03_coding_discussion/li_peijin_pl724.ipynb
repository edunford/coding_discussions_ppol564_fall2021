```python
import pandas as pd
import numpy as np 
```

```python
crime_data = pd.read_csv('chicago_summer_2018_crime_data.csv')
crime_data.head()
```

```python
##select the data we need
data_count=crime_data.filter(["primary_type","month","day"])
data_count.head()
```

```python
##calculate the number of different crime types happending each day
a=data_count.groupby(["primary_type","month","day"]).day.size().reset_index(name = 'count')
a.head()
```

```python
##transfer the df from long-to-wide and calculate the total number of each day's crime
data_count2=a.pivot_table(values="count",columns="day",index=["primary_type","month"],fill_value = 0)
data_count2['Total']=  data_count2.sum(axis=1)
data_count2.head()
```

```python
##present the proportion of each day's crime 
for x in range(len(data_count2)):
    data_count2.iloc[x]=round(data_count2.iloc[x]/data_count2.iloc[x, -1], 2)
data_count2.sort_values(['primary_type','month'])## for readable 
data_count2
```

```python
## drop the last column of total number 
data_count2.drop(columns=["Total"])
```
